```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/student_data.csv')
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.columns
```

```python
d={'Gender':'Gender', 'EthnicGroup':'Group', 'ParentEduc':'ParentED', 'LunchType':'LunchType', 'TestPrep':'TestPrep',
       'ParentMaritalStatus':'ParentsMaritalStatus', 'PracticeSport':'PracticeSport', 'IsFirstChild':'IsFirstChild', 'NrSiblings':'NoOfSibilings',
       'TransportMeans':'Transport', 'WklyStudyHours':'Studyhours/week', 'MathScore':'MathScore', 'ReadingScore':'ReadingScore',
       'WritingScore':'WritingScore'}
df.rename(columns=d,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['NoOfSibilings']=df['NoOfSibilings'].fillna(0.0)
```

```python
df['ParentED']=df['ParentED'].fillna('Non-Disclosed')
```

```python
df['TestPrep']=df['TestPrep'].fillna('not-mentioned')
```

```python
df['ParentsMaritalStatus']=df['ParentsMaritalStatus'].fillna('Non-Disclosed')
```

```python
df['IsFirstChild']=df['IsFirstChild'].fillna('Non-Disclosed')
```

```python
df['Transport']=df['Transport'].fillna('yet to assign')
```

```python
df['Studyhours/week']=df['Studyhours/week'].fillna('Non-Disclosed')
```

```python
df['Group']=df['Group'].fillna('To be Assigned')
```

```python
df['PracticeSport']=df['PracticeSport'].fillna('Not-Disclosed')
```
